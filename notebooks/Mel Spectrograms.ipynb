```python
import librosa
import librosa.display
import IPython.display as ipd
import matplotlib.pyplot as plt
```

# Loading audio files with Librosa

```python
prefix = "../songs/"
cfm_file = prefix + "cfm.wav"
fancy_file  = prefix + "fancy.wav"
pw_file = prefix + "pw.wav"

cfm, sr = librosa.load(cfm_file)
fancy, _ = librosa.load(fancy_file)
pw, _ = librosa.load(pw_file)
```

# Mel filter banks

```python
filter_banks = librosa.filters.mel(n_fft=2048, sr=22050, n_mels=10)
```

```python
filter_banks.shape
```

```python
plt.figure(figsize=(25, 10))
librosa.display.specshow(filter_banks, 
                         sr=sr, 
                         x_axis="linear")
plt.colorbar(format="%+2.f")
plt.show()
```


# Extracting Mel Spectrogram

```python
mel_spectrogram_cfm = librosa.feature.melspectrogram(cfm, sr=sr, n_fft=2048, hop_length=512, n_mels=60)
```

```python
mel_spectrogram_cfm.shape
```

```python
log_mel_spectrogram_cfm = librosa.power_to_db(mel_spectrogram_cfm)
```

```python
log_mel_spectrogram_cfm.shape
```

```python
plt.figure(figsize=(25, 10))
librosa.display.specshow(log_mel_spectrogram_cfm, 
                         x_axis="time",
                         y_axis="mel", 
                         sr=sr)
plt.colorbar(format="%+2.f")
plt.show()
```

```python
mel_spectrogram_fancy = librosa.feature.melspectrogram(fancy, sr=sr, n_fft=2048, hop_length=512, n_mels=60)
log_mel_spectrogram_fancy = librosa.power_to_db(mel_spectrogram_fancy)
plt.figure(figsize=(25, 10))
librosa.display.specshow(log_mel_spectrogram_fancy, 
                         x_axis="time",
                         y_axis="mel", 
                         sr=sr)
plt.colorbar(format="%+2.f")
plt.show()
```

```python
mel_spectrogram_pw = librosa.feature.melspectrogram(pw, sr=sr, n_fft=2048, hop_length=512, n_mels=60)
log_mel_spectrogram_pw = librosa.power_to_db(mel_spectrogram_pw)
plt.figure(figsize=(25, 10))
librosa.display.specshow(log_mel_spectrogram_pw, 
                         x_axis="time",
                         y_axis="mel", 
                         sr=sr)
plt.colorbar(format="%+2.f")
plt.show()
```
