```python
import librosa
import librosa.display
import IPython.display as ipd
import matplotlib.pyplot as plt
```

```python
prefix = "../songs/"
cfm_file = prefix + "cfm.wav"
fancy_file  = prefix + "fancy.wav"
pw_file = prefix + "pw.wav"

cfm, sr = librosa.load(cfm_file)
fancy, _ = librosa.load(fancy_file)
pw, _ = librosa.load(pw_file)
```

```python
mfccs_cfm = librosa.feature.mfcc(cfm, n_mfcc=13, sr=sr)
mfccs_fancy = librosa.feature.mfcc(fancy, n_mfcc=13, sr=sr)
mfccs_pw = librosa.feature.mfcc(pw, n_mfcc=13, sr=sr)
```

```python
mfccs_cfm.shape
```

```python
# Visualize MFCCs

plt.figure(figsize=(25, 10))
librosa.display.specshow(mfccs_cfm, 
                         x_axis="time", 
                         sr=sr)
plt.colorbar(format="%+2.f")
plt.show()
```

```python

plt.figure(figsize=(25, 10))
librosa.display.specshow(mfccs_fancy, 
                         x_axis="time", 
                         sr=sr)
plt.colorbar(format="%+2.f")
plt.show()
```

```python

plt.figure(figsize=(25, 10))
librosa.display.specshow(mfccs_pw, 
                         x_axis="time", 
                         sr=sr)
plt.colorbar(format="%+2.f")
plt.show()
```

```python
delta_mfccs_cfm = librosa.feature.delta(mfccs_cfm)
delta2_mfccs_cfm = librosa.feature.delta(mfccs_cfm, order=2)


delta_mfccs_fancy = librosa.feature.delta(mfccs_fancy)
delta2_mfccs_fancy = librosa.feature.delta(mfccs_fancy, order=2)


delta_mfccs_pw = librosa.feature.delta(mfccs_pw)
delta2_mfccs_pw = librosa.feature.delta(mfccs_pw, order=2)
```

```python
plt.figure(figsize=(25, 10))
librosa.display.specshow(delta_mfccs_cfm, 
                         x_axis="time", 
                         sr=sr)
plt.colorbar(format="%+2.f")
plt.show()
```

```python
plt.figure(figsize=(25, 10))
librosa.display.specshow(delta2_mfccs_cfm, 
                         x_axis="time", 
                         sr=sr)
plt.colorbar(format="%+2.f")
plt.show()
```

```python
plt.figure(figsize=(25, 10))
librosa.display.specshow(delta_mfccs_fancy, 
                         x_axis="time", 
                         sr=sr)
plt.colorbar(format="%+2.f")
plt.show()
```

```python
plt.figure(figsize=(25, 10))
librosa.display.specshow(delta2_mfccs_cfm, 
                         x_axis="time", 
                         sr=sr)
plt.colorbar(format="%+2.f")
plt.show()
```

```python
plt.figure(figsize=(25, 10))
librosa.display.specshow(delta_mfccs_pw, 
                         x_axis="time", 
                         sr=sr)
plt.colorbar(format="%+2.f")
plt.show()
```

```python
plt.figure(figsize=(25, 10))
librosa.display.specshow(delta2_mfccs_pw, 
                         x_axis="time", 
                         sr=sr)
plt.colorbar(format="%+2.f")
plt.show()
```
