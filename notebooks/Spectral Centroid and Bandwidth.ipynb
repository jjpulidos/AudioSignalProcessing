```python
import math
import matplotlib.pyplot as plt
import numpy as np
import librosa
import librosa.display
import IPython.display as ipd
```

```python
prefix = "../songs/"
cfm_file = prefix + "cfm.wav"
fancy_file  = prefix + "fancy.wav"
pw_file = prefix + "pw.wav"

cfm, sr = librosa.load(cfm_file, duration=50, offset=5)
fancy, _ = librosa.load(fancy_file, duration=50, offset=5)
pw, _ = librosa.load(pw_file, duration=50, offset=5)
```

```python
# Spectral centroid


FRAME_SIZE = 1024
HOP_LENGTH = 512

sc_cfm = librosa.feature.spectral_centroid(y=cfm, sr=sr, n_fft=FRAME_SIZE, hop_length=HOP_LENGTH)[0]
sc_fancy = librosa.feature.spectral_centroid(y=fancy, sr=sr, n_fft=FRAME_SIZE, hop_length=HOP_LENGTH)[0]
sc_pw = librosa.feature.spectral_centroid(y=pw, sr=sr, n_fft=FRAME_SIZE, hop_length=HOP_LENGTH)[0]
```

```python
frames_cfm = range(len(sc_cfm))
t_cfm = librosa.frames_to_time(frames_cfm, hop_length=HOP_LENGTH)

frames_fancy = range(len(sc_fancy))
t_fancy = librosa.frames_to_time(frames_fancy, hop_length=HOP_LENGTH)

frames_pw = range(len(sc_pw))
t_pw = librosa.frames_to_time(frames_pw, hop_length=HOP_LENGTH)
```

```python

plt.figure(figsize=(25,10))

plt.plot(t_cfm, sc_cfm, color='b')
plt.plot(t_fancy, sc_fancy, color='r')
plt.plot(t_pw, sc_pw, color='y')

plt.show()
```

```python
# Spectral Bandwidth

sb_cfm = librosa.feature.spectral_bandwidth(y=cfm, sr=sr, n_fft=FRAME_SIZE, hop_length=HOP_LENGTH)[0]
sb_fancy = librosa.feature.spectral_bandwidth(y=fancy, sr=sr, n_fft=FRAME_SIZE, hop_length=HOP_LENGTH)[0]
sb_pw = librosa.feature.spectral_bandwidth(y=pw, sr=sr, n_fft=FRAME_SIZE, hop_length=HOP_LENGTH)[0]
```

```python

plt.figure(figsize=(25,10))

plt.plot(t_cfm, sb_cfm, color='b')
plt.plot(t_fancy, sb_fancy, color='r')
plt.plot(t_pw, sb_pw, color='y')

plt.show()
```
